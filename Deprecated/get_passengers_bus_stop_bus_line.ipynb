Aggregate trip data

```python
travel_data = travel_data.\
    drop_duplicates().\
    groupby(['DAY_TYPE', 'ORIGIN_PT_CODE', 'DESTINATION_PT_CODE']).\
    agg({'TOTAL_TRIPS': 'sum'}).\
    reset_index()
    
```

Example Case:

Suppose there are 4 bus stops in a bus line in the following order: [BS_1, BS_2, BS_3, BS_4], where the bus line refers to a specific bus service and direction. Suppose also the number of trips are as follows:

BS_1 -> BS_4 = 1
BS_1 -> BS_3 = 2
BS_1 -> BS_2 = 3

BS_2 -> BS_3 = 4
BS_2 -> BS_4 = 5

BS_3 -> BS_4 = 6

Note that the data above implies that the passengers are distinct. Therrefore, there are a total of
(1+2+...+6) = 21 passengers.

Then the number of passengers at BS_1 for this bus line = 1+2+3 = 6
Then the number of passengers at BS_2 for this bus line = 1+2+3+4+5 = 15
Then the number of passengers at BS_3 for this bus line = 1+2+4+5+6 = 18 #Mote that we do not add 3
Then the number of passengers at BS_4 for this bus line = 6

Our goal is to get these numbers but abstract it out for our use case.

So, given a bus service, and direction

$$
\text{Number of passengers at } BS_i = \sum_{\text{travel}} \left( \text{number of passengers} \mid \text{origin sequence} \geq i \text{ and } \text{destination sequence} \leq i \right)
$$
