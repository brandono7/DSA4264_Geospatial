```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the data (adjust the paths as necessary)
bus_stops_df = pd.read_csv('bus_stops.csv')  # Assuming this is already loaded properly
mrt_stations_df = pd.read_csv('mrt_stations_with_geo_data.csv')

# Create a figure using Plotly Graph Objects
fig = go.Figure()

# Add MRT stations to the map
fig.add_trace(go.Scattermapbox(
    lat=mrt_stations_df['latitude'],
    lon=mrt_stations_df['longitude'],
    mode='markers',
    marker=go.scattermapbox.Marker(size=9, color='red'),
    text=mrt_stations_df['station_name'],
    name='MRT Stations'  # This will add the MRT stations to the legend
))

# Add bus stops to the map
fig.add_trace(go.Scattermapbox(
    lat=bus_stops_df['Latitude'],
    lon=bus_stops_df['Longitude'],
    mode='markers',
    marker=go.scattermapbox.Marker(size=7, color='blue'),
    text=bus_stops_df['Description'],
    name='Bus Stops'  # This will add the bus stops to the legend
))

# Set up the layout for the map
fig.update_layout(
    title='MRT Stations and Bus Stops in Singapore',
    mapbox=dict(
        style="carto-positron",
        zoom=11,
        center=dict(lat=mrt_stations_df['latitude'].mean(), lon=mrt_stations_df['longitude'].mean())
    ),
    margin={"r":0,"t":50,"l":0,"b":0},
    showlegend=True  # This ensures that the legend is shown on the map
)

# Display the figure
fig.show()
```
